```python

r = requests.get(
    'https://www.quandl.com/api/v3/datasets/FSE/AFX_X?start_date=2017-01-01&end_date=2017-12-31&api_key=HvCsSjxYKRQWEXcxJum-') 
#DELETE KEY BEFORE MAKING PUBLIC

overview_2017 = r.json()
data_2017 = overview_2017['dataset']['data']
keys = overview_2017['dataset']['column_names']

#create dict with keys as keys and data_2017 as values: days
days = {}
for d in data_2017:
    d = dict(zip(keys, d)) #changes each day entry into a dict
    days[d['Date']] = d
    
#Calculate what the highest and lowest opening prices were for the stock in this period: highs, lows, and date_keys
highs = []
lows = []
date_keys = []
for date, stats in days.items():
    highs.append(stats['High'])
    lows.append(stats['Low'])
    date_keys.append(date)

#find max and min and their respective dates: max_high, highest_date, min_low, lowest_date
max_high = max(highs)
highest_date = date_keys[highs.index(max_high)]
print('THE HIGHEST PRICE POINT OF THE STOCK: ' + str(max_high) + ' AT ' + str(highest_date))
min_low = min(lows)
lowest_date = date_keys[lows.index(min_low)]
print('THE LOWEST PRICE POINT OF THE STOCK: ' + str(min_low) + ' AT ' + str(lowest_date))

#find largest change in any one day (make sure you use the absolute value)
intraday_changes = [abs(high-low) for high, low in zip(highs, lows)]
biggest_change = max(intraday_changes)
print('THE BIGGEST PRICE CHANGE WITHIN A DAY: ' + str(biggest_change) + ' AT ' + str(date_keys[intraday_changes.index(biggest_change)]))
#^ the long float is ugly but may be useful, also try to print whether the change was a drop or a rise

#find largest change in any two days based on closing price (make sure abolute value): closes, two_day_deltas, twodaydeltas_max, change_date_index
closing_prices = []
[closing_prices.append(stats['Close']) for date, stats in days.items()]
two_day_deltas = []
for index, d in enumerate(closing_prices):
    if index == 0:
        continue
    delta = d - closing_prices[index-1]
    two_day_deltas.append(abs(delta))
twodaydeltas_max = max(two_day_deltas)
change_date_index = two_day_deltas.index(twodaydeltas_max) #index of the date of the highest change
print('THE LARGEST CHANGE WITHIN ANY TWO DAYS: ' + str(twodaydeltas_max) + ' BETWEEN ' + 
      str(date_keys[change_date_index]) + ' & ' + str(date_keys[change_date_index-1]))


#find average daily trading volume (ADTV) of the year: ADTV
daily_trading_vols = []
for key, value in days.items():
    daily_trading_vols.append(value['Traded Volume'])
total_trading_volume = sum(daily_trading_vols)
ADTV = total_trading_volume / len(days)
print('AVERAGE DAILY TRADING VOLUME: ' + str(round(ADTV)))
#find median daily trading volume during the year: median_DTV
median_DTV = st.median(daily_trading_vols)
print('MEDIAN DAILY TRADING VOLUME: ' + str(median_DTV))

```
